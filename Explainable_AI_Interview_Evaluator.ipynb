```python
import re
import gradio as gr
from transformers import pipeline
```

```python
nlp = pipeline(
    "text-generation",
    model="gpt2",
    max_new_tokens=120
)
```

```python
def analyze_interview(transcript):
    text = transcript.lower()

    # ---------- Filler Word Analysis ----------
    filler_words = ["um", "uh", "like", "you know", "actually", "basically"]
    filler_count = sum(text.count(word) for word in filler_words)

    # ---------- Communication Score ----------
    sentence_count = len(re.findall(r'[.!?]', transcript))
    word_count = len(transcript.split())

    communication_score = min(
        100,
        int((sentence_count / max(word_count, 1)) * 1000)
    )

    # ---------- Confidence Score ----------
    weak_phrases = ["i think", "maybe", "not sure", "i guess"]
    weak_count = sum(text.count(w) for w in weak_phrases)

    confidence_score = max(
        0,
        100 - (weak_count * 10) - (filler_count * 5)
    )

    # ---------- Strengths ----------
    strengths = []
    if communication_score > 60:
        strengths.append("Clear sentence structure")
    if confidence_score > 60:
        strengths.append("Confident tone")
    if filler_count < 3:
        strengths.append("Minimal filler words")

    if not strengths:
        strengths.append("Shows basic communication ability")

    # ---------- AI Suggestions ----------
    prompt = f"""
    The following is an interview response.
    Give 3 short improvement tips for better interview performance.

    Response:
    {transcript}
    """

    suggestions = nlp(prompt)[0]["generated_text"]

    bias_feedback = bias_and_generic_analysis(transcript)

    return (
        f"{communication_score}/100",
        f"{confidence_score}/100",
        filler_count,
        ", ".join(strengths),
        bias_feedback,
        suggestions
    )
```

```python
interface = gr.Interface(
    fn=analyze_interview,
    inputs=gr.Textbox(
        label="Paste Interview Transcript",
        lines=8,
        placeholder="Paste candidate interview answer here..."
    ),
    outputs=[
    gr.Textbox(label="Communication Score"),
    gr.Textbox(label="Confidence Score"),
    gr.Number(label="Filler Word Count"),
    gr.Textbox(label="Strengths"),
    gr.Textbox(label="Bias & Generic Language Analysis"),
    gr.Textbox(label="Improvement Suggestions")
]
,
    title="🎤 Explainable AI Interview Evaluator",
    description="AI-powered evaluation of interview responses with explainable feedback"
)

interface.launch()
```

```python
def bias_and_generic_analysis(transcript):
    text = transcript.lower()

    generic_phrases = [
        "hardworking", "passionate", "team player",
        "quick learner", "self motivated", "dedicated"
    ]

    weak_phrases = [
        "i think", "maybe", "i guess", "not sure", "probably"
    ]

    generic_count = sum(text.count(p) for p in generic_phrases)
    weak_count = sum(text.count(p) for p in weak_phrases)

    bias_warnings = []

    if generic_count > 1:
        bias_warnings.append(
            "Uses generic self-descriptive terms without evidence."
        )

    if weak_count > 1:
        bias_warnings.append(
            "Shows uncertainty that may reduce perceived confidence."
        )

    if not bias_warnings:
        bias_warnings.append(
            "No major bias or generic language issues detected."
        )

    return " ".join(bias_warnings)
```
